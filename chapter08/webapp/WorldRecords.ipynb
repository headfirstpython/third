```python
class EventCountError(Exception):
    """Raised if the excepted world record data is not complete/correct."""
    pass
```

```python
## RECORDS = (0, 2, 4, 5)   # The original table numbers from the printed book.

RECORDS = (0, 1, 3, 4)   # Update these values as required (these are valid as of Wednesday 28, 2026).

COURSES = ("LC Men", "LC Women", "SC Men", "SC Women")
```

```python
list(zip(RECORDS, COURSES))
```

```python

# Something will go wrong if the RECORDS values are incorrect (perhaps due to someone editing the
# Wikipedia webpage). Each course should have (at least) 20 entries, so the code in this cell has 
# been adjusted to raise an error if the event_count is less that this.

for table, course in zip(RECORDS, COURSES):
    event_count = 0
    print(f"{course}:")
    for n, row in enumerate(tables[table].find("tr", mode="all")[1:], 1):
        columns = row.find("td", mode="all")
        event = columns[0].text
        time = columns[1].text
        print(f"{n}\t{event} -> {time}")
        event_count = event_count + 1
    if event_count < 20:
        raise EventCountError("Please check the values in the RECORDS tuple (data missing).")
    print()
    
```
